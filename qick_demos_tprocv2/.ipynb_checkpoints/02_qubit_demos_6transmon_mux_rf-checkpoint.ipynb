# Configure the RF board and channels for the additional loopback pair (no fridge)

```python
### MUX TEST
TEST_GEN_CH=5 # Qubit DAC channel 
RO_CH=0
RO_CH_RF = 4 # New variable that we need for QICK box since
# according to https://github.com/openquantumhardware/qick/wiki/RF216-(RF-board-for-ZCU216) the RO ADC is RF board channel 4.

res_center_freq = 6330 # To be in the middle of the res freqs. 3000-5000 see nothing,6000 and 7000 see something, 8000+ see nothing 
soc.rfb_set_gen_filter(TEST_GEN_CH, fc=res_center_freq/1000, ftype='bandpass', bw=1.0)
soc.rfb_set_ro_filter(RO_CH_RF, fc=res_center_freq/1000, ftype='bandpass', bw=1.0)
# Set attenuator on DAC.
soc.rfb_set_gen_rf(TEST_GEN_CH, 30, 27) # Verified 30->25 see increased gain in loopback
# Set attenuator on ADC.
soc.rfb_set_ro_rf(RO_CH_RF, 30) # Verified 30->25 see increased gain in loopback
```

# TOF Measurement

Measures the "time of flight" of the measurement pulse. The time of flight is the time at which the measurement pulse appears in the ADC buffer. We only want to start capturing data from this point in time onwards. We store this time of flight in the parameter <code> config["trig_time"] </code>.

```python
class LoopbackProgram(AveragerProgramV2):
    def _initialize(self, cfg):
        self.ro_ch = cfg['ro_chs']
        self.gen_ch = cfg['gen_ch']
        
        self.declare_gen(ch=self.gen_ch, nqz=cfg['nqz'], ro_ch=self.ro_ch, mixer_freq=cfg['mixer_freq'])

        # tproc configured
        self.declare_readout(ch=self.ro_ch, length=cfg['ro_len'])
        self.add_readoutconfig(ch=self.ro_ch, name="myro", freq=cfg['pulse_freqs'], gen_ch=self.gen_ch)

        self.add_pulse(ch=self.gen_ch, name="myconst", ro_ch=self.ro_ch, 
                       style="const", 
                       length=cfg["pulse_len"],
                       freq=cfg['pulse_freqs'], 
                       phase=cfg['pulse_phases'],
                       gain=cfg['pulse_gains'],
                      )
    
    def _body(self, cfg):
        self.send_readoutconfig(ch=self.ro_ch, name="myro", t=0)
        self.pulse(ch=self.gen_ch, name="myconst", t=0)
        self.trigger(ros=[self.ro_ch], pins=[0], t=0, ddr4=True)

        
config = {'gen_ch': TEST_GEN_CH,
          'ro_chs': RO_CH,
          'mixer_freq': 6000,
          'nqz': 2,
          'pulse_freqs': 6200,
          'pulse_gains': 0.9,
          'pulse_phases': 0.0,
          'ro_phases': 0.0,
          'trig_time': 0.0,
          'pulse_len': 0.5,
          'ro_len': 1.0,
         }
print(config)
prog = LoopbackProgram(soccfg, reps=1, final_delay=0.5, cfg=config)
iq_list = prog.acquire_decimated(soc, soft_avgs=10)
```

```python
# Plot results.
t = prog.get_time_axis(ro_index=0)
plt.plot(t, iq_list[0][:,0], label="I value")
plt.plot(t, iq_list[0][:,1], label="Q value")
plt.plot(t, np.abs(iq_list[0].dot([1,1j])), label="magnitude")
plt.legend()
plt.ylabel("ADC Units")
plt.xlabel("us");
plt.title('Time of Flight Experiment- Loopback, no fridge')
######## SET THIS VALUE AS YOUR config['trig_time'] #########
print("max magnitude =", np.max(np.abs(iq_list[0].dot([1,1j]))))

# Trim DAC attenuator
# 30, 30 1 is 3000
# 30, 20 is 5000
# 20, 20 is 5000 (clipped)

# Trim ADC attenuator
# 30, 20 is 3000 (clipped)

# Now all signals are smaller after we did 0, 0 on the DAC one time :( 

# 30 30 30, 0.5, 300
# 30 30 30, 1, 600
# 30 30 27, 1, 800 but very noisy
# 30 27 30, 1, 800 but cleaner
# 27 27 30, 1, 1000 but very noisy, something is compressing

# These numbers are good for running with the fridge: 30, 27, 30
```

```python
# Verify that if you set the DAC channel to 4 or 6 you see nothing in loopback (Verified)

# Therefore we know that for this setup, DAC 5 is connected to ADC 0 (RF channel 4)
```

```python
# Convert max magnitude into dBm and see if that roughly matches expectation for a 6200 MHz loopback
# Gain of 0.01 corresponds roughly to  with -30 dBm attenuation DAC side + 15 dB attenuation ADC side (which means +75 dB amplification) 
# Gain of 1 corresponds roughly to 0 dBm out of the RFSoC so gain of 0.01 corresponds to around -30 dBm out of the RFSoC.

# -30 - (15+15) = -60 dB + 60 dBm = 0 dBm which corresponds to 1254 units on the ADC 

# -30 - (15+21) = -66 dB + 60 dBm = -6 dBm which corresponds to 692 units on the ADC

# If I set the DAC to be at 30+20 dB attenuation then I have -30-50 dBm + 75 dB = -5 dBm which corresponds to ~200 units on the ADC
```

# Configure the RF board and channels for the 6 qubit system in loopback

```python
### MUX TEST
GEN_CH=6 # Qubit DAC channel 
MIXMUXGEN_CH = 4 # Resonator DAC channel
MUXRO_CH = [2,3,4,5,6,7] # Resonator readout channels
MUXRO_CH_RF = 5 # New variable that we need for QICK box since
# according to https://github.com/openquantumhardware/qick/wiki/RF216-(RF-board-for-ZCU216) the MUX RO ADC is RF board channel 5.

qubit_center_freq = 4400 # To be in the middle of the qubit freqs.
res_center_freq = 6330 # To be in the middle of the res freqs. 3000-5000 see nothing,6000 and 7000 see something, 8000+ see nothing 
soc.rfb_set_gen_filter(MIXMUXGEN_CH, fc=res_center_freq/1000, ftype='bandpass', bw=1.0)
soc.rfb_set_gen_filter(GEN_CH, fc=qubit_center_freq/1000, ftype='bandpass', bw=1.0)
soc.rfb_set_ro_filter(MUXRO_CH_RF, fc=res_center_freq/1000, ftype='bandpass', bw=1.0)
# Set attenuator on DAC.
soc.rfb_set_gen_rf(MIXMUXGEN_CH, 2, 10) # Verified 30->25 see increased gain in loopback
soc.rfb_set_gen_rf(GEN_CH, 15, 15) # Verified 30->25 see increased gain in loopback
# Set attenuator on ADC.
soc.rfb_set_ro_rf(MUXRO_CH_RF, 0) # Verified 30->25 see increased gain in loopback
```

# TOF Measurement

Measures the "time of flight" of the measurement pulse. The time of flight is the time at which the measurement pulse appears in the ADC buffer. We only want to start capturing data from this point in time onwards. We store this time of flight in the parameter <code> config["trig_time"] </code>.

```python
class MuxProgram(AveragerProgramV2):
    def _initialize(self, cfg):
        ro_chs = cfg['ro_chs']
        gen_ch = cfg['gen_ch']
        
        self.declare_gen(ch=gen_ch, nqz=cfg['nqz'], ro_ch=ro_chs[0], 
                         mux_freqs=cfg['pulse_freqs'], 
                         mux_gains=cfg['pulse_gains'], 
                         mux_phases=cfg['pulse_phases'],
                         mixer_freq=cfg['mixer_freq'])
        for ch, f, ph in zip(cfg['ro_chs'], cfg['pulse_freqs'], cfg['ro_phases']):
            self.declare_readout(ch=ch, length=cfg['ro_len'], freq=f, phase=ph, gen_ch=gen_ch)

        self.add_pulse(ch=gen_ch, name="mymux", 
                       style="const", 
                       length=cfg["pulse_len"],
                       mask=[0,1,2,3,4,5],
                      )
        
    def _body(self, cfg):
        self.trigger(ros=cfg['ro_chs'], pins=[0], t=cfg['trig_time'], ddr4=True)
        self.pulse(ch=cfg['gen_ch'], name="mymux", t=0)
        
config = {'gen_ch': MIXMUXGEN_CH,
          'ro_chs': MUXRO_CH,
          'mixer_freq': 6000,
          'nqz': 2,
          'pulse_freqs': [6191.519, 6216, 6292.321, 6405.85, 6432.959, 6468.441],
          'pulse_gains': [1.0]*6,
          'pulse_phases': [0.0]*6,
          'ro_phases': [0.0]*6,
          'trig_time': 0.0,
          'pulse_len': 0.5,
          'ro_len': 2.0,
         }
print(config)
prog = MuxProgram(soccfg, reps=1, final_delay=0.5, cfg=config)
iq_list = prog.acquire_decimated(soc, soft_avgs=300)
```

```python
t = prog.get_time_axis(ro_index=0)
fig, axes = plt.subplots(len(config['ro_chs']), 1, figsize=(12,12))
for i, ch in enumerate(config['ro_chs']):
    plot = axes[i]
    plot.plot(t, iq_list[i][:,0], label="I value")
    plot.plot(t, iq_list[i][:,1], label="Q value")
    plot.plot(t, np.abs(iq_list[i].dot([1,1j])), label="magnitude")
    plot.legend()
    plot.set_ylabel("a.u.")
    plot.set_xlabel("us");
    ######## SET THIS VALUE AS YOUR config['trig_time'] #########
    plot.axvline(0.75, c='r') 
```
